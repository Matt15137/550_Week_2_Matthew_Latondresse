```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset 
data = pd.read_csv('C:/Users/mattl/OneDrive/Documents/GitHub/DSC550/Largest_Companies.csv')

```

```python
# Introduction
print("Introduction:")
print("This dataset contains information about the largest companies in terms of revenue.")
print("It includes details such as company names, industries, revenue, revenue growth, employees, and headquarters locations.")

```

```python
# Questions to Explore
print("\nQuestions to Explore:")
print("1. What are the top industries represented in the dataset?")
print("2. Is there a correlation between revenue and the number of employees?")
print("3. How does the revenue growth vary among different industries?")

```

```python
# Clean the 'Revenue growth' column by removing '%' and converting to numeric
data['Revenue growth'] = data['Revenue growth'].str.replace('%', '').astype(float)

# Graph 1: Top Industries Represented
top_industries = data['Industry'].value_counts().nlargest(10)
plt.figure(figsize=(10, 6))
top_industries.plot(kind='bar')
plt.title("Top Industries Represented in the Dataset")
plt.xlabel("Industry")
plt.ylabel("Number of Companies")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Graph 2: Correlation between Revenue and Employees
plt.figure(figsize=(8, 8))
plt.scatter(data['Revenue (USD millions)'], data['Employees'], alpha=0.5)
plt.title("Correlation between Revenue and Number of Employees")
plt.xlabel("Revenue (USD millions)")
plt.ylabel("Number of Employees")
plt.grid(True)
plt.tight_layout()
plt.show()

# Graph 3: Revenue Growth by Industry
industry_growth = data.groupby('Industry')['Revenue growth'].mean().nlargest(10)
industry_growth.plot(kind='bar', figsize=(12, 6))
plt.title("Average Revenue Growth by Industry")
plt.xlabel("Industry")
plt.ylabel("Average Revenue Growth")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
# Conclusions
print("\nConclusions:")
print("1. The dataset covers a variety of industries, with some industries having more representation than others.")
print("2. There is a positive correlation between revenue and the number of employees, suggesting that larger companies tend to have more employees.")
print("3. The average revenue growth varies among industries, with certain industries experiencing higher growth rates than others.")

```
